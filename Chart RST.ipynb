```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn import preprocessing

df = pd.read_excel (r'chart.xlsx')
df.sort_values('Investments',inplace=True, ascending=False)
```

```python
t=df.Tech.values

x=df.Patents.values
y=df.Publications.values

c=df.Type.values

#c - colors
c=np.where(c=='Плохой', 'red', c)
c=np.where(c=='Устойчивый', 'blue', c)
c=np.where(c=='Перспективный', 'green', c)

#s - circle size

s=df.Investments.values
s_normalized=preprocessing.normalize([s])
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(16.5, 12)

for i, txt in enumerate(t):
    ax.annotate(
            txt, 
            xy=(x[i], y[i]),
            xytext=(x[i], y[i]),
            color=c[i],
            arrowprops=dict(facecolor=c[i]),
            horizontalalignment='left',
            verticalalignment='bottom')

ax.set_xlim([-0.1, 0.71])
ax.set_ylim([-0.15, 0.32])
plt.xlabel('Патенты')
plt.ylabel('Публикации')
plt.title('Инвестиции в технологии в 2017 г')

ax.scatter(x, y, s/10, color =c,  linewidths=0.6, edgecolors='blue', alpha=0.5)
plt.grid(True)
plt.show()
fig.savefig('left_bottom.png', dpi=100)
```

```python

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.xlabel('Патенты')
plt.ylabel('Публикации')
plt.title('Инвестиции в технологии в 2017 г')

plt.scatter(x, y, s/10, color =c,  linewidths=0.6, edgecolors='blue', alpha=0.5)
plt.grid(True)
plt.show()
fig.savefig('test2png.png', dpi=100)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(18.5, 10.5)

y_pos = np.arange(len(t))

ax.barh(y_pos, s, xerr=x*100, align='center',
        color='green', ecolor='black', alpha=0.5)
ax.set_yticks(y_pos)
ax.set_yticklabels(t)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Инвестиции')
ax.set_title('Инвестиции в технологии в 2017 г')

plt.show()
```
